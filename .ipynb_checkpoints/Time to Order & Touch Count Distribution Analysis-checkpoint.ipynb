```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt

#display options
import warnings
warnings.filterwarnings('ignore')
sns.set(rc={'image.cmap': 'coolwarm'})
pd.set_option('display.max_columns', 500)
%config InlineBackend.figure_format = "retina"
plt.style.use("fivethirtyeight")
```

```python
df = pd.read_csv('time_to_order.csv')
```

```python
df.head()
```

```python
def describe(df):
    return pd.concat([df[['touch_count','time_to_order']].describe().T,
                      df[['touch_count','time_to_order']].median().rename('median'),
                      df[['touch_count','time_to_order']].mad().rename('mad')
                     ], axis=1).T

describe(df)
```
